```python
#Importing Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
```

```python
#dropping columns not needed
df_data =df_data.drop(['Status','Year'], axis=1)
df_data.head()
```

```python
#grouping of data by country
df_data = df_data.groupby('Country').mean()
df_data.head()
```

```python
#seperating features and label
df_labels = df_data['Life expectancy ']
df_features = df_data.drop('Life expectancy ', axis = 1)
```

```python
#number of null values in each coumn
df_features.isnull().sum()
```

```python
#replacing null values by mean of each column
df_features.fillna(value = df_features.mean(), inplace = True)
df_labels.fillna(value = df_labels.mean(), inplace = True)
```

```python
#splitting of data for training and testing
df_features_train, df_features_test, df_labels_train, df_labels_test = train_test_split(
        df_features, df_labels, train_size = 0.7, test_size = 0.3)
#fitting the model
lasso=Lasso(alpha=0,max_iter=10)
lasso.fit(df_features_train,df_labels_train)
lasso.predict(df_features_test)
```

```python
lasso.score(df_features_train, df_labels_train)
```

```python
lasso.score(df_features_test, df_labels_test)
```
